# Inter-rater Reliability

Here we measure IRR of a given datasets: KEMDy19 and KEMDy20. IRR, as the name insists, represents the relibability between raters. There are many ways to measure this and we have used the following.
* Regression: Valence & arousal
  - Pearson correlation
* Classification: Emotion and valence & arousal (5-level rated)
  - Fleiss' Kappa
  - Krippendorff Alpha

We analyze this is to point out the gap between given datasets and for future works on this.

```python
from erc.interrater import kemdy19_get_rater, kemdy20_get_rater, plot_heatmap, get_irr

kemdy19 = kemdy19_get_rater()
kemdy20 = kemdy20_get_rater()
```
