```python
from collections import defaultdict

import pandas as pd
```

```python
fname = "wandb_export_2023-04-13T11_30_20.490+09_00.csv"
df1 = pd.read_csv(fname)
fname = "scratch_onehot-concat_peakl.csv"
df2 = pd.read_csv(fname)

df = pd.concat([df1, df2], axis=1)
df
```

```python
measures = [
    "epoch/valid_acc",
    "epoch/valid_macrof1",
    "epoch/valid_microf1",
    "epoch/valid_ccc(aro)",
    "epoch/valid_ccc(val)",
]

runs = [
    "Scratch | peakl",
    "Scratch | soft",
    "Scratch | one-hot",
    "Concat | peakl",
]
```

```python
df.filter(like="Scratch | soft | val=4").filter(like="epoch/valid_acc").iloc[:, 1].max()
```

```python
total = defaultdict(dict)
for run in runs:
    total[run] = defaultdict(list)
    for val_fold in range(5):
        run_name = f"{run} | val={val_fold}"
        print(run_name)
        _df = df.filter(like=run_name)
        for measure in measures:
            m = _df.filter(like=measure).iloc[:, 1].max()
            total[run][measure].append(m)
```

```python
for run in runs:
    print(run)
    data = pd.DataFrame(total[run])
    for metric, m, s in zip(data.columns, data.mean().values, data.std().values):
        print(f"{metric}: {m:.3f} ± {s:.3f}")
```

```python
pd.DataFrame(total['Scratch | peakl']).mean()
```

```python
import pickle
with open("fold_result.pkl", mode="wb") as f:
    pickle.dump(obj=total, file=f)
```
