```python
import ast
import numpy as np 
from pingouin import ttest
from statannot import add_stat_annotation 

peakl = df.iloc[0]
soft = df.iloc[1]
hard = df.iloc[2]


for metric in ['epoch/valid_acc', 'epoch/valid_macrof1','epoch/valid_microf1', 'epoch/valid_ccc(aro)', 'epoch/valid_ccc(val)']:
    # metric = 'epoch/valid_acc'

    epoch = np.array([i for i in range(1,5+1)])
    peakl_array = np.array(ast.literal_eval(peakl[metric]))
    soft_array = np.array(ast.literal_eval(soft[metric]))
    hard_array = np.array(ast.literal_eval(hard[metric]))

    result_df = pd.DataFrame()
    result_df['Subject'] = epoch
    result_df['PeakL'] = peakl_array
    result_df['Soft'] = soft_array
    result_df['Hard'] = hard_array
    ax = result_df.drop('Subject',axis=1).plot(kind="box", title=metric,)
    add_stat_annotation(ax, data = result_df,

                                        box_pairs =[("PeakL","Soft"), 
                                                    ("PeakL","Hard"),
                                                    ("Soft","Hard")],

                                        test ='t-test_paired',
                                        text_format ='star',
                                        loc ='inside')
    plt.show()
```
