```python
import numpy as np 
import pandas as pd 

def split_str(x:str):
    ppeakle = 'use_peakl=False'

    if x.count('|') == 5:
        # label_type = 'onehot'
        model_name, fold_num, label_type, rdeuce_, ppeakle, batch_ = x.split('|')
    elif x.count('|') == 4:
        model_name, fold_num, label_type, rdeuce_, ppeakle = x.split('|')
    elif x.count('|') == 3:
        model_name, fold_num, label_type, batch_ = x.split('|')
    else:
        model_name, fold_num, label_type = x.split('|')

    return pd.Series([model_name.strip(), fold_num.strip(), label_type.strip(), ppeakle.strip()])

FILE_PATH ="/home/hoesungryu/etri-erc/wandb_export_2023-04-07T20_05_07.959+09_00.csv"

SELECTED_COLS = ['Name',
 'epoch/valid_acc',
 'epoch/valid_auroc',
 'epoch/valid_ccc(aro)',
 'epoch/valid_ccc(val)',
 'epoch/valid_cls_loss',
 'epoch/valid_f1',
 'epoch/valid_loss',
 'epoch/valid_reg_loss',
]

SELECTED_COLS_2 = ['Name',
 'step/train_acc',
 'step/train_auroc',
 'step/train_ccc(aro)',
 'step/train_ccc(val)',
 'step/train_cls_loss',
 'step/train_f1',
 'step/train_loss',
 'step/train_reg_loss',
 'step/valid_acc',
 'step/valid_auroc',
 'step/valid_ccc(aro)',
 'step/valid_ccc(val)',
 'step/valid_cls_loss',
 'step/valid_f1',
 'step/valid_loss',
 'step/valid_reg_loss',
]
```

```python
df = pd.read_csv(FILE_PATH)[SELECTED_COLS]
df[["model_name","fold_num","label_type",'ppeakle']] = df['Name'].apply(split_str)

for model_name in ['RobertaL', 'Hub_RobertaL']:
    print(f"{model_name}\n")
    tmp_onehot = df[(df['model_name'] == model_name) & (df['label_type'] == 'one-hot')]
    tmp_multi = df[(df['model_name'] == model_name) & (df['label_type'] == 'multi') & (df['ppeakle'] == "use_peakl=False")]
    tmp_multi_ppeakl = df[(df['model_name'] == model_name) & (df['label_type'] == 'multi') & (df['ppeakle'] == "use_peakl=True")]
    for metric in ['acc','f1','ccc(aro)','ccc(val)']:
        print(f"\t{metric}")
        print("\t\tonehot: {:.4f}+/-{:.4f}".format(tmp_onehot[f'epoch/valid_{metric}'].mean(), tmp_onehot[f'epoch/valid_{metric}'].std()))
        print("\t\tmulti: {:.4f}+/-{:.4f}".format(tmp_multi[f'epoch/valid_{metric}'].mean(), tmp_multi[f'epoch/valid_{metric}'].std()))
        print("\t\tmulti_ppeakl: {:.4f}+/-{:.4f}".format(tmp_multi_ppeakl[f'epoch/valid_{metric}'].mean(), tmp_multi_ppeakl[f'epoch/valid_{metric}'].std()))
    print(' ')
```
